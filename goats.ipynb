```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# ============================================================================
# SETTINGS / CONSTANTS
# ============================================================================

def weighted_land_requirement(zones):
    """
    Calculate weighted land requirement per animal across multiple zones.
    
    zones: list of dicts, each with keys:
        - 'proportion': fraction of total land in this zone
        - 'land_per_goat': average ha/goat for the zone
    """
    return sum(zone['proportion'] * zone['land_per_goat'] for zone in zones)


def carrying_capacity(total_land_ha, land_per_goat):
    """
    Calculate carrying capacity given total land and land per goat.
    
    total_land_ha: total rangeland area in hectares
    land_per_goat: weighted land requirement per goat
    """
    return total_land_ha / land_per_goat


# Input parameters
total_rangeland_ha = 27_405_600  # 27,405,600 ha

# Weighted land requirements per rainfall zone
arid_semi_arid = weighted_land_requirement([
    {'proportion': 0.5, 'land_per_goat': 0.75},   # 200–400 mm
    {'proportion': 0.5, 'land_per_goat': 0.45},   # 400–612.5 mm
])

high_rainfall = 0.195  # ha/goat for >612.5 mm

# Combine weighted zones
overall_land_per_goat = 0.8 * arid_semi_arid + 0.2 * high_rainfall

# Calculate carrying capacity
K = carrying_capacity(total_rangeland_ha, overall_land_per_goat)
print(f"Estimated carrying capacity of goats: {int(K):,}")

def compute_quarterly_offtake_multipliers(
        low_season=153.9,
        baseline=177.9,
        festive=244.0,
        q4_override=1.10):
    """
    Returns quarterly multipliers Q1–Q4 as indices 0–3.
    Q1 = low-season ratio
    Q2 = high-season (festive) ratio
    Q3 = baseline normal
    Q4 = user override (e.g., 1.10)
    """
    q1 = low_season / baseline
    q2 = festive / baseline
    q3 = 1.0
    q4 = q4_override

    return {
        0: q1,
        1: q2,
        2: q3,
        3: q4
    }
```

```python
# ============================================================================
# INPUT / BASE PARAMETERS (ANNUAL)
# ============================================================================

# Historical herd data from KNBS
HISTORICAL_YEARS = [2018, 2019, 2020, 2021, 2022, 2023, 2024]
HISTORICAL_TOTALS = [25876113, 34657969, 35358674, 33123213, 29965055, 34385998, 37776633]
HERD0 = HISTORICAL_TOTALS[-1]

# Demographic / reproductive parameters
SEX_RATIO = 0.50                    # 0.48-0.52
LITTER_SIZE = 1.50                  # 1.50-1.97
KIDDING_RATE = 1.10                 # 0.51-1.68

# Mortality; aggregated quarterly below
KID_MORTALITY = 0.24                # 0.24-0.45
WEANER_MORTALITY = 0.09             # 0.07-0.15
GROWER_MORTALITY = 0.06             # 0.05-0.10
OVERALL_ADULT_MORTALITY = 0.05      # 0.05-0.10

# Offtake rates
OFFTAKE_RATE_GROWERS = 0.18         # 0.15-0.30
OFFTAKE_RATE_DOES = 0.10            # 0.20-0.40
OFFTAKE_RATE_BUCKS = 0.35           # 0.30-0.80
OFFTAKE_Q_MULT = compute_quarterly_offtake_multipliers()

# Culling / replacement
CULL_RATE_DOES_ANNUAL = 0.04        # 0.03-0.07
CULL_RATE_BUCKS_ANNUAL = 0.03       # 0.02-0.05

# Productivity / weights
PROD_PARAMS = {
    # 'growers': {'carcass_yield_pct': 0.48, 'avg_live_weight_kg': 30.0},  # carcass 0.43-0.50
    'does': {
        'carcass_yield_pct': 0.48,                                       # 0.45-0.52
        'avg_live_weight_kg': 38.0,                                      # 35-45
        'lactating_fraction': 0.22,                                      # 0.15-0.30
        'milk_yield_l_per_day': 0.40,                                    # 0.32-1.92
        'lactating_days': 200                                            # 142-284
    },
    'bucks': {'carcass_yield_pct': 0.50, 'avg_live_weight_kg': 48.0}     # bucks carcass 0.48-0.55
}

# Herd composition targets
HISTORICAL_PROPORTIONS = {
    'kids': 0.18,     # 0.15-0.25
    'weaners': 0.10,  # 0.08-0.15
    'growers': 0.10,  # 0.08-0.15
    'does': 0.56,     # 0.45-0.60
    'bucks': 0.06     # 0.02-0.15
}

TARGET_BUCK_PROP = 0.07  # 0.05-0.10 (soft cap)

# DMI / feed params
FEED_PARAMS = {
    'kids': {'dmi_pct': 0.035, 'kid_w': 3.0, 'weaner_w': 22.0},            # DMI kids 0.030-0.040
    'weaners': {'dmi_pct': 0.04, 'weaner_w': 22.0, 'grower_w': 30.0},      # DMI weaners 0.035-0.050
    'growers': {'dmi_pct': 0.045, 'grower_w': 30.0},                       # DMI growers 0.040-0.050
    'does': {'dmi_pct': 0.023, 'doe_w': 38.0, 'lact_mult': 1.5},           # DMI does 0.020-0.025
    'bucks': {'dmi_pct': 0.023, 'buck_w': 50.0}                            # DMI bucks 0.020-0.025
}

# Seasonal multipliers
KIDDING_SEASONALITY_Q = [1.00, 1.08, 0.95, 1.06]         # Q1..Q4
KID_MORTALITY_SEASONALITY_Q = [1.04, 0.92, 1.00, 0.97]

# Transition probabilities
P_KID_TO_WEANER = 0.55    # 0.45-0.66
P_WEANER_TO_GROWER = 0.50 # 0.40-0.60
P_GROWER_TO_ADULT = 0.30  # 0.20-0.40

# Forecast horizon
FORECAST_YEARS = 10
T_QUARTERS = FORECAST_YEARS * 4
```

```python
# ============================================================================
# HELPERS
# ============================================================================

def initialize_herd(total_herd):
    kids = int(total_herd * HISTORICAL_PROPORTIONS['kids'])
    weaners = int(total_herd * HISTORICAL_PROPORTIONS['weaners'])
    growers = int(total_herd * HISTORICAL_PROPORTIONS['growers'])
    does = int(total_herd * HISTORICAL_PROPORTIONS['does'])
    bucks = int(total_herd * HISTORICAL_PROPORTIONS['bucks'])
    return {
        'kids': float(kids),
        'weaners': float(weaners),
        'growers': float(growers),
        'does': float(does),
        'bucks': float(bucks)
    }

def quarterly_rate_linear(annual_rate):
    return annual_rate / 4.0

def compute_effective_rates_linear(t, forage_index):
    q = t % 4
    f = forage_index - 1.0

    kidding_q = quarterly_rate_linear(KIDDING_RATE) * KIDDING_SEASONALITY_Q[q] * (1 + 0.05 * f)

    kid_m_q = quarterly_rate_linear(KID_MORTALITY) * KID_MORTALITY_SEASONALITY_Q[q] * max(0.001, 1 - 0.02 * f)
    weaner_m_q = quarterly_rate_linear(WEANER_MORTALITY)
    grower_m_q = quarterly_rate_linear(GROWER_MORTALITY)
    adult_m_q = quarterly_rate_linear(OVERALL_ADULT_MORTALITY)

    dm = OFFTAKE_Q_MULT[q]

    off_g_q = quarterly_rate_linear(OFFTAKE_RATE_GROWERS) * dm
    off_d_q = quarterly_rate_linear(OFFTAKE_RATE_DOES) * dm
    off_b_q = quarterly_rate_linear(OFFTAKE_RATE_BUCKS) * dm

    cull_d_q = quarterly_rate_linear(CULL_RATE_DOES_ANNUAL)
    cull_b_q = quarterly_rate_linear(CULL_RATE_BUCKS_ANNUAL)

    return {
        'kidding_q': kidding_q,
        'kid_mort_q': kid_m_q,
        'weaner_mort_q': weaner_m_q,
        'grower_mort_q': grower_m_q,
        'adult_mort_q': adult_m_q,
        'offtake_g_q': off_g_q,
        'offtake_d_q': off_d_q,
        'offtake_b_q': off_b_q,
        'cull_d_q': cull_d_q,
        'cull_b_q': cull_b_q
    }
```

```python
def quarter_meat_and_milk_counts(state, rates):
    # growers = state['growers']
    does = state['does']
    bucks = state['bucks']

    # only count MARKET offtake as meat (do NOT include culls)
    # off_g = growers * rates['offtake_g_q']
    off_d = does * rates['offtake_d_q']
    off_b = bucks * rates['offtake_b_q']

    # gp = PROD_PARAMS['growers']
    dp = PROD_PARAMS['does']
    bp = PROD_PARAMS['bucks']

    meat = (
        # off_g * gp['avg_live_weight_kg'] * gp['carcass_yield_pct'] +
        off_d * dp['avg_live_weight_kg'] * dp['carcass_yield_pct'] +
        off_b * bp['avg_live_weight_kg'] * bp['carcass_yield_pct']
    )

    lact = does * dp['lactating_fraction']
    milk = lact * dp['milk_yield_l_per_day'] * (dp['lactating_days'] / 4.0)

    return meat, milk
```

```python
# ============================================================================
# FEED
# ============================================================================
def compute_feed_intake(state):
    p = FEED_PARAMS
    kids = state['kids'] * p['kids']['dmi_pct'] * (p['kids']['weaner_w'] - p['kids']['kid_w'])
    weaners = state['weaners'] * p['weaners']['dmi_pct'] * (p['weaners']['grower_w'] - p['weaners']['weaner_w'])
    growers = state['growers'] * p['growers']['dmi_pct'] * p['growers']['grower_w']
    does = state['does'] * p['does']['dmi_pct'] * p['does']['doe_w']
    lact_does = state['does'] * p['does']['lact_mult'] * p['does']['dmi_pct'] * p['does']['doe_w']
    bucks = state['bucks'] * p['bucks']['dmi_pct'] * p['bucks']['buck_w']
    return kids + weaners + growers + does + lact_does + bucks
```

```python
# ============================================================================
# SIMULATION
# ============================================================================
def simulate_forecast(initial_state):
    state = initial_state.copy()
    rec = []

    for t in range(T_QUARTERS):
        q = t % 4
        forage_idx = [1.00, 1.02, 0.98, 1.01][q]
        rates = compute_effective_rates_linear(t, forage_idx)

        # # production for reporting
        # meat, milk = quarter_meat_and_milk_counts(state, rates)

        ks = state['kids']
        ws = state['weaners']
        gs = state['growers']
        ds = state['does']
        bs = state['bucks']
        total = ks + ws + gs + ds + bs

        # Placeholder capacity factors (recalculated after transitions)
        cap_birth = 1.0
        cap_mature = max(0.85, 1.0 - 0.25 * (total / K))

        # Apply mortality (survivors this quarter)
        # Add density-dependent mortality increase
        density_mort_factor = 1.0 + 0.5 * max(0, (total / K) - 0.8)
        
        ks_m = ks * (1.0 - rates['kid_mort_q'] * density_mort_factor)
        ws_m = ws * (1.0 - rates['weaner_mort_q'] * density_mort_factor)
        gs_m = gs * (1.0 - rates['grower_mort_q'] * density_mort_factor)
        ds_m = ds * (1.0 - rates['adult_mort_q'] * density_mort_factor)
        bs_m = bs * (1.0 - rates['adult_mort_q'] * density_mort_factor)

        # Offtake (remove animals for meat/trade)
        ks_o = ks_m                                  # we assume negligible targeted kid offtake before weaning here
        ws_o = ws_m * (1.0 - rates['offtake_g_q'])
        gs_o = gs_m * (1.0 - rates['offtake_g_q'])
        ds_o = ds_m * (1.0 - rates['offtake_d_q'])
        bs_o = bs_m * (1.0 - rates['offtake_b_q'])

        # Culling
        ds_c = ds_o * (1.0 - rates['cull_d_q'])
        bs_c = bs_o * (1.0 - rates['cull_b_q'])
        gs_c = gs_o  # no explicit cull on growers in this config

        # Transitions through life stages
        # kids -> weaners
        mk = ks_o * P_KID_TO_WEANER
        ks_n = ks_o - mk
        ws_i = ws_o + mk

        # weaners -> growers
        mw = ws_i * P_WEANER_TO_GROWER
        ws_n = ws_i - mw
        gs_i = gs_c + mw

        # growers -> adults (dampened by cap_mature)
        mg = gs_i * P_GROWER_TO_ADULT * cap_mature
        gs_n = gs_i - mg

        # split matured into does and bucks by sex ratio
        new_d = mg * SEX_RATIO
        new_b = mg * (1.0 - SEX_RATIO)

        ds_n = ds_c + new_d
        bs_n = bs_c + new_b

        # Enforce buck proportion soft cap
        tentative_total = ks_n + ws_n + gs_n + ds_n + bs_n
        max_allowed_bucks = TARGET_BUCK_PROP * max(1.0, tentative_total)
        if bs_n > max_allowed_bucks:
            excess_bucks = bs_n - max_allowed_bucks
            bs_n = max_allowed_bucks

        # Recalculate capacity factor using projected population after transitions
        projected_pre_birth = ks_n + ws_n + gs_n + ds_n + bs_n
        cap_birth = max(0.3, 1.0 - 0.7 * (projected_pre_birth / K))

        # Births: use recalculated cap_birth
        # effective_kidding_rate_q = rates['kidding_q'] * FERTILITY_RATE * FECUNDITY_RATE * (1.0 - ABORTION_RATE)
        # births = ds_n * effective_kidding_rate_q * LITTER_SIZE * cap_birth
        births = ds_n * rates['kidding_q'] * LITTER_SIZE * cap_birth
        ks_n += births

        state = {'kids': ks_n, 'weaners': ws_n, 'growers': gs_n, 'does': ds_n, 'bucks': bs_n}
        total_after = sum(state.values())

        # production for reporting
        meat, milk = quarter_meat_and_milk_counts(state, rates)

        feed = compute_feed_intake(state)

        rec.append({
            'year': 2025 + t // 4,
            'quarter': q + 1,
            'kids': ks_n,
            'weaners': ws_n,
            'growers': gs_n,
            'does': ds_n,
            'bucks': bs_n,
            'total_start': total,
            'meat_kg': meat,
            'milk_l': milk,
            'feed_intake': feed,
            'total_after': total_after
        })

    return pd.DataFrame(rec)
```

```python
# ============================================================================
# HISTORICAL
# ============================================================================
historical_df = pd.DataFrame({'year': HISTORICAL_YEARS, 'total': HISTORICAL_TOTALS})
for c, prop in HISTORICAL_PROPORTIONS.items():
    historical_df[c] = (historical_df['total'] * prop).astype(int)

def compute_historical_production(df, params):
    out = []
    for _, r in df.iterrows():
        kids = r['kids']
        weaners = r['weaners']
        growers = r['growers']
        does = r['does']
        bucks = r['bucks']
        meat_y = 0.0
        milk_y = 0.0
        for q in range(4):
            rates = compute_effective_rates_linear(q, 1.0)
            s = {'kids': kids/4.0, 'weaners': weaners/4.0, 'growers': growers/4.0, 'does': does, 'bucks': bucks}
            m, ml = quarter_meat_and_milk_counts(s, rates)
            meat_y += m
            milk_y += ml
        out.append({'year': r['year'], 'meat_kg': meat_y, 'milk_l': milk_y})
    return pd.DataFrame(out)

historical_prod_df = compute_historical_production(historical_df, PROD_PARAMS)
```

```python
# ============================================================================
# RUN
# ============================================================================
initial_state = initialize_herd(HERD0)
forecast_df = simulate_forecast(initial_state)

forecast_annual = forecast_df.groupby('year').agg(
    {
        'meat_kg': 'sum',
        'milk_l': 'sum',
        'total_start': 'mean',
        'feed_intake': 'sum'
    }
).reset_index().rename(columns={'total_start': 'total_mean_start'})
```

```python
# ============================================================================
# CAGR CALCULATION
# ============================================================================

def cagr(start, end, periods):
    return (end / start) ** (1 / periods) - 1


# ----------------------------
# HISTORICAL CAGR (2018–2024)
# ----------------------------
hist_years = 2024 - 2018

hist_total_cagr = cagr(
    historical_df.loc[historical_df.year == 2018, 'total'].values[0],
    historical_df.loc[historical_df.year == 2024, 'total'].values[0],
    hist_years
)

hist_meat_cagr = cagr(
    historical_prod_df.loc[historical_prod_df.year == 2018, 'meat_kg'].values[0],
    historical_prod_df.loc[historical_prod_df.year == 2024, 'meat_kg'].values[0],
    hist_years
)

hist_milk_cagr = cagr(
    historical_prod_df.loc[historical_prod_df.year == 2018, 'milk_l'].values[0],
    historical_prod_df.loc[historical_prod_df.year == 2024, 'milk_l'].values[0],
    hist_years
)

# feed not modeled historically → NaN placeholder
hist_feed_cagr = np.nan


# ----------------------------
# FORECAST CAGR (2025–2034)
# ----------------------------
fc_years = 2034 - 2025

forecast_start = forecast_annual.loc[forecast_annual.year == 2025]
forecast_end = forecast_annual.loc[forecast_annual.year == 2034]

fc_total_cagr = cagr(
    forecast_start['total_mean_start'].values[0],
    forecast_end['total_mean_start'].values[0],
    fc_years
)

fc_meat_cagr = cagr(
    forecast_start['meat_kg'].values[0],
    forecast_end['meat_kg'].values[0],
    fc_years
)

fc_milk_cagr = cagr(
    forecast_start['milk_l'].values[0],
    forecast_end['milk_l'].values[0],
    fc_years
)

fc_feed_cagr = cagr(
    forecast_start['feed_intake'].values[0],
    forecast_end['feed_intake'].values[0],
    fc_years
)


# ----------------------------
# OUTPUT
# ----------------------------
cagr_df = pd.DataFrame({
    'metric': ['Total Population', 'Meat', 'Milk', 'Feed'],
    'historical_cagr_2018_2024': [
        hist_total_cagr,
        hist_meat_cagr,
        hist_milk_cagr,
        hist_feed_cagr
    ],
    'forecast_cagr_2025_2034': [
        fc_total_cagr,
        fc_meat_cagr,
        fc_milk_cagr,
        fc_feed_cagr
    ]
})

print(cagr_df)
```
